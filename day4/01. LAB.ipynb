```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2',' A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
}, index=[0,1,2,3])
df1
```

```python
df2 = pd.DataFrame({
    'A': ['A4', 'A5', 'A6',' A7'],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'C': ['C4', 'C5', 'C6', 'C7'],
    'D': ['D4', 'D5', 'D6', 'D7']
}, index=[4,5,6,7])
df2
```

```python
df3 = pd.DataFrame({
    'A': ['A8', 'A9', 'A10',' A11'],
    'B': ['B8', 'B9', 'B10', 'B11'],
    'C': ['C8', 'C9', 'C10', 'C11'],
    'D': ['D8', 'D9', 'D10', 'D11']
}, index=[8,9,10,11])
df3
```

```python
pd.concat([df1, df2, df3])
```

```python
pd.concat([df1, df2, df3], keys=['X','Y','Z'])
```

```python
left = pd.DataFrame({
    'key' : ['K1', 'K2', 'K3'],
    'lval' : [1, 2, 3] 
})
left
```

```python
right = pd.DataFrame({
    'key' : ['K0', 'K1', 'K3'],
    'rval' : [4, 5, 6]
})
right
```

```python
pd.merge(left, right, on='key')
```

```python
pd.merge(left, right, on='key', how='right')
```

```python
pd.merge(left, right, on='key', how='outer')
```

```python
pd.merge(left, right, on='key', how='inner')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2019', periods=1000))
ts
```

```python
ts = ts.cumsum()
```

```python
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000,4), index=ts.index, columns=['A','B','C','D'])
df
```

```python
df = df.cumsum()
```

```python
plt.figure()
df.plot()
plt.legend(loc='best')
```

```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
# HTML 분석하기
soup = BeautifulSoup(res, "html.parser")
```

```python
data = soup.select_one('div.head_info > span.value').string
print(data)
```

```python
soup.select('div.head_info > span.value')
```
